```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.offline as ply
import plotly.graph_objs as go
ply.init_notebook_mode(connected=True)
```

```python
file = "/home/adam/Documents/vidas/data/KITTI/data_tracking_oxts/training/oxts/0000.txt"
```

```python
df = pd.read_csv(file, sep=" ", header=None).drop(columns=[30])
df.head()
```

```python
headers = ['lat',
         'lon',
         'alt',
         'roll',
         'pitch',
         'yaw',
         'vn',
         've',
         'vf',
         'vl',
         'vu',
         'ax',
         'ay',
         'az',
         'af',
         'al',
         'au',
         'wx',
         'wy',
         'wz',
         'wf',
         'wl',
         'wu',
         'posacc',
         'velacc',
         'navstat',
         'numsats',
         'posmode',
         'velmode',
          "orimode"
          ]
df.columns = headers
```

```python
df
```

```python
# position
ply.iplot(dict(
    data=[go.Scatter3d(x=df.lat, y=df.lon, z=df.alt, mode="markers",
                     marker=dict(color=df.posacc, showscale=True, size=1, colorbar=dict(title="Position Error")))],
    layout=dict(
        scene=dict(aspectratio=dict(x=1,y=1,z=0.2))
    )
))
```

```python
# velocity
ply.iplot(dict(
    data=[go.Scatter3d(x=df.ve, y=df.vn, z=df.vu, mode="markers",
                     marker=dict(color=df.velacc, showscale=True, size=1, colorbar=dict(title="Velocity Error")))],
    layout=dict(
        scene=dict(aspectmode="data")
    )
))
```

```python
# Speed
euclidean = lambda xs: sum(x**2 for x in xs) ** 0.5
speed = df[["ve", "vn", "vu"]].apply(euclidean, axis=1)

ply.iplot(dict(
    data=[go.Scatter(y=speed, mode="markers", 
                     error_y=dict(type="data", array=df.velacc),
                     marker=dict(color=df.velacc, showscale=True, colorbar=dict(title="Velocity Error")))],
    layout=dict(
        height=700,
        width=700,
    )
))
```

```python
# navigation modes
df.iloc[:, -5:].plot()
```

```python
# position/velocity
ply.iplot(dict(
    data=[go.Cone(x=df.lat, y=df.lon, z=df.alt, u=df.vn, v=df.ve, w=df.vu,
                 sizemode="absolute", sizeref=10, 
                  colorbar=dict(title=dict(text="Speed"))),
          go.Scatter3d(x=df.lat, y=df.lon, z=df.alt, mode="lines", opacity=0.5, 
                       line=dict(width=5))
         
         ],
    layout=dict(
        scene=dict(aspectratio=dict(x=1,y=1,z=0.2))
    )
))
```

```python
# position/acceleration
ply.iplot(dict(
    data=[go.Cone(x=df.lat, y=df.lon, z=df.alt, u=df.ax, v=df.ay, w=df.az,
                 sizemode="absolute", sizeref=30, 
                  colorbar=dict(title=dict(text="Acceleration"))),
#           go.Scatter3d(x=df.lat, y=df.lon, z=df.alt, mode="lines", opacity=0.5, 
#                        line=dict(width=5))
         
         ],
    layout=dict(
        scene=dict(aspectratio=dict(x=1,y=1,z=0.2))
    )
))
```

```python
rel_acc = df[["vn", "ve", "vu"]].apply(euclidean, axis=1).diff()

# position/acceleration
ply.iplot(dict(
    data=[
#         go.Cone(x=df.lat, y=df.lon, z=df.alt, u=df.ax, v=df.ay, w=df.az,
#                  sizemode="absolute", sizeref=30, 
#                   colorbar=dict(title=dict(text="Acceleration"))),
          go.Scatter3d(x=df.lat, y=df.lon, z=df.alt, mode="lines", opacity=0.5, 
                  line=dict(width=5, color=rel_acc, colorbar=dict(title="Acceleration")))
         
         ],
    layout=dict(
        scene=dict(aspectratio=dict(x=1,y=1,z=0.2))
    )
))
```

```python
# accuracy vs num satellites
ply.iplot(dict(
    data=[
        go.Scatter(
            x=df.numsats,
            y=df.velacc,
            mode="markers",
            name="velacc"
            ),
        go.Scatter(
            x=df.numsats,
            y=df.posacc,
            mode="markers",
            name="posacc"
            ),
    ]
))
```
